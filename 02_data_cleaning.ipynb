```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/SDM/Desktop/Titanic_Capstone/Data/Titanic-Dataset.csv")
df.head()
```

```python
## Remove unnecessary columns not useful for survival prediction

df.drop(columns = ["PassengerId", "Ticket"], inplace = True)
```

```python
## Now Handle Missing Values (Age) - It fills missing Age values using the median age of people with the same Pclass and Sex.

df['Age'] = df.groupby(['Pclass', 'Sex'])['Age'].transform(
    lambda x: x.fillna(x.median()))
```

```python
## Filling the missing values in the Embarked column with the most common embarkation port.(.mode()[0] Takes the first most frequent value)

df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
## Places 1 who has cabin and 0 where there is no canin by creating new column
df['Has_Cabin'] = df['Cabin'].apply(lambda x: 0 if pd.isna(x) else 1)
df.drop(columns='Cabin', inplace=True)
```

```python
## Verifying if data is free of null value

df.isnull().sum()
```

```python
df.to_csv("outputs/titanic_cleaned.csv", index=False)
```

## Data Cleaning Summary

- Dropped irrelevant columns such as PassengerId and Ticket
- Handled missing values in Age using median by class and gender
- Filled missing Embarked values using mode
- Converted Cabin into a binary feature indicating availability
- Resulting dataset contains no missing values
