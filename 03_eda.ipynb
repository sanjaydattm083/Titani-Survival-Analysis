```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_style("whitegrid")
```

```python
df = pd.read_csv("outputs/titanic_cleaned.csv")
df.head()
```

```python
##Survival Count
df['Survived'].value_counts()
```

```python
##Survival Plot
sns.countplot(x='Survived', data=df)
plt.title("Survival Distribution, Observation: More passengers did not survive")
plt.show()
```

```python
##Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender, Females had a much higher survival rate than males")
plt.show()
```

```python
##Survival by Passenger Class

sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class, Higher-class passengers had better survival chances")
plt.show()
```

```python
## Survivors generally paid higher fares
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title("Fare vs Survival")
plt.show()

```

```python
##Correlation Heatmap
numeric_df = df.select_dtypes(include='number')

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='viridis')
plt.title("Feature Correlation")
plt.show()
```

## Key EDA Insights

- Female passengers had significantly higher survival rates

- First-class passengers were more likely to survive
 
- Higher fare and cabin availability positively influenced survival

- Socio-economic status played a major role in survival outcomes
